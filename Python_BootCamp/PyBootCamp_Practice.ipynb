```python
def extractEvenReverse(dataAr): 
    '''Extracts even values from array and reverses it 
    arguments:
    dataAr: numpy array
    '''
    if not isinstance(dataAr,np.array):
        print ("you should have passed an array")
        return -1
    return dataAr[dataAr % 2 == 0][::-1]
```
