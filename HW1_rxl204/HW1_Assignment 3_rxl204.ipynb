### Generate 100 2-dimensional random datapoints in gaussian distribution

```python
ReprRand = np.random.randn(2,100)
```

```python
ReprRand.shape
```

### Generate 50 more random 2x100 numpy arrays and store them in ReprRandAll, all with different mean

```python
ReprRandAll = np.zeros((50, 2, 100))
for i in range(50): 
    ReprRandAll[i] = np.random.randint(1,10) * np.random.randn(2,100) + np.random.randint(1,10)
```

```python
ReprRandAll.shape
```

### Plot points

```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
%pylab inline

for i in range(50):
    plt.plot(ReprRandAll[i,0], ReprRandAll[i,1],'o')
plt.show()
```
